```python
listaEliminarNa = ['Odometro', 
                   'Velocidad', 
                   'HorasMotor', 
                   'ControlCrucero', 
                   'TemperaturaMotor', 
                   'RPM', 
                   'NivelFuel',
                   'ConsumoTotal',
                   'Frenadas',
                   'TiempoRalenti',
                   'ParMotor',
                   'Pedal'
                  ]
```

Vamos a imputar los Nan de **Embrague**, interpolando con datos conocidos, podemos verlo en el siguienyte enlace:

No las ha borrado, probamos con otro método.

```python
# Eliminamos filas con NAN, en Embrague que es en la única que quedan
dfEnd = dfEnd[~np.isnan(dfEnd).any(axis=1)]
```

```python
dfEnd.isnull().sum()
```

```python
dfTime.isnull().sum()
```

Vamos a pasar la mayoría de variables a tipo **int** ya que float ocupa mucha memoria y da problemas de memoria valga la redundancia. Dejamos **ConsumoTotal** que va realizando anotaciones cada 0.5, y como será nuestra variable objetivo pues para que sea más fidedigna, también dejamos como float **Odometro**..

```python
columnas = ['Velocidad', 'TemperaturaMotor', 'Frenadas', 'TiempoRalenti', 'ParMotor', 'Pedal', 'Embrague', 'RPM', 
            'ControlCrucero', 'NivelFuel', 'HorasMotor']
```

```python
for c in columnas:
    dfEnd[c] = dfEnd[c].astype(int)
```
